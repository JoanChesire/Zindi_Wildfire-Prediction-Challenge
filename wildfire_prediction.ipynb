# Data spliting 


```python
# Create a copy of the train_df so that we maintain data integrity of the original 

train_copy=train_df.copy().dropna()
```

 Do not split the data randomly - this would give us artificially high scores. Instead, let's use the last 3 years of the dataset for validation to more closely match the test configuration.

```python
train = train_copy.loc[train_copy.date < '2011-01-01']
valid = train_copy.loc[train_copy.date > '2011-01-01']
print(train.shape, valid.shape)
```

```python
train.info()
```

# Models

```python
# Define input and output columns - you can play with adding or removing inputs to the model
X_train=train.drop(columns=['burn_area','ID','month','year','lat','lon','date'])
y_train=train['burn_area']
X_valid=valid.drop(columns=['burn_area','ID','month','year','lat','lon','date'])
y_valid=valid['burn_area']
```

#### Linear Regression using RidgeCV


```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error,r2_score

# Create and fit the model
model=RidgeCV()
model.fit(X_train, y_train)

# Make predictions
preds = model.predict(X_valid)

# Score
mse= mean_squared_error(y_valid, preds)
r2=r2_score(y_valid,preds)
rmse= mean_squared_error(y_valid, preds)**0.5 
print("mse:",mse)
print("r2_score:",r2)
print("rmse:",rmse)
```

#### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=10,max_depth=5,random_state=42)

#fit the model
model.fit(X_train,y_train)

preds=model.predict(X_valid)

#score
mse= mean_squared_error(y_valid,preds)
r2=r2_score(y_valid,preds)
rmse= mean_squared_error(y_valid, preds)**0.5

print('mse ofRandomForestRegressor:',mse)
print("r2_score of RandomForestRegressor:",r2_score)
print("rmse of RandomForestRegressor:",rmse)

```

```python
# Ensure the test features match the training features
test_features = test_df.drop(columns=['ID','month','year','lat','lon','date'])  # Drop the ID column

# Make predictions on the test data
test_preds = model.predict((test_features).fillna(0))

```

```python

#  submission DataFrame with ID and predicted burn_area
submission = pd.DataFrame({
    'ID': test_df['ID'],
    'burn_area': test_preds
})

# Save the submission DataFrame to a CSV file
submission.to_csv('submission.csv', index=False)
```

## Gradient Boosting Regressor
The gradient boosting regressor builds the trees sequentially

```python
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
gmodel=GradientBoostingRegressor(n_estimators=10,learning_rate=0.1,max_depth=3,random_state=42)
gmodel.fit(X_train,y_train)

preds=gmodel.predict(X_valid)

#score
mse= mean_squared_error(y_valid,preds)
r2=r2_score(y_valid,preds)
rmse= np.sqrt(mse)

print('mse ofRandomForestRegressor:',mse)
print("r2_score of RandomForestRegressor:",r2_score)
print("rmse of RandomForestRegressor:",rmse)
```

- The gradient boosting regressor has a higher `RMSE` compared to the random forest regressor
