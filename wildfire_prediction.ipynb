```python
# Heat map with burn_area as the target variable
import seaborn as sns
corr_matrix=train_df.select_dtypes(include=['number']).corr()


plt.figure(figsize=(8,12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1,center=0)
plt.title('Correlation Heatmap with Emphasis on Burn Area')
plt.show()

burn_area_corr = corr_matrix[['burn_area']].sort_values(by='burn_area', ascending=False)

plt.figure(figsize=(8, 12))
sns.heatmap(burn_area_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation with Burn Area')
plt.show()
```

### Adding Features 

```python
# Check the  date column part of the data frame which is placed in the ID part of the data frame
train_df['ID'].dtypes
print(train_df['ID'])
```

```python
train_df['date']=pd.to_datetime(train_df['ID'].apply(lambda x: x.split('_')[1]))
test_df['date']=pd.to_datetime(test_df['ID'].apply(lambda x: x.split('_')[1]))
```

```python
train_df['date']
```

```python
train_df.head()
```

```python
# create the month and year columns
train_df['month']=train_df.date.dt.month
test_df['month']=test_df.date.dt.month
train_df['year']=train_df.date.dt.year
test_df['year']=test_df.date.dt.year
```

```python
train_df.head()
```

```python
#Group by month and calculate the mean burn_area
monthly_mean_burn_area = train_df.groupby('month')['burn_area'].mean().reset_index()
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.bar(monthly_mean_burn_area['month'], monthly_mean_burn_area['burn_area'], color='skyblue')
plt.xlabel('Month')
plt.ylabel('Mean Burn Area')
plt.title('Mean Burn Area by Month')
plt.xticks(monthly_mean_burn_area['month'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Group by year and get the mean yearly burn area
yearly_mean_burn_area=train_df.groupby('year')['burn_area'].mean().reset_index()
```

```python
#barplot of the year burn area
plt.figure(figsize=(10,6))
plt.bar(yearly_mean_burn_area['year'],yearly_mean_burn_area['burn_area'],color='skyblue')
plt.xlabel('year')
plt.ylabel('yearly_mean_burn_area')
plt.title('Mean burn area by Year')
plt.xticks(yearly_mean_burn_area['year'])
plt.grid(axis='y',linestyle='--',alpha=0.8)
plt.show()
```
