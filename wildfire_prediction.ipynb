```python
# Create a copy of the train_df so that we maintain data integrity of the original 

train_copy=train_df.copy().dropna()
```

 Do not split the data randomly - this would give us artificially high scores. Instead, let's use the last 3 years of the dataset for validation to more closely match the test configuration.

```python
train = train_copy.loc[train_copy.date < '2011-01-01']
valid = train_copy.loc[train_copy.date > '2011-01-01']
print(train.shape, valid.shape)
```

# Models

```python
# Define input and output columns - you can play with adding or removing inputs to the model
X_train=train.drop(columns=['burn_area','ID','month','year','lat','lon','date'])
y_train=train['burn_area']
X_valid=valid.drop(columns=['burn_area','ID','month','year','lat','lon','date'])
y_valid=valid['burn_area']
```

#### Linear Regression using RidgeCV


```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error,r2_score

# Create and fit the model
model=RidgeCV()
model.fit(X_train, y_train)

# Make predictions
preds = model.predict(X_valid)

# Score
mse= mean_squared_error(y_valid, preds)
r2=r2_score(y_valid,preds)
rmse= mean_squared_error(y_valid, preds)**0.5 
print("mse:",mse)
print("r2_score:",r2)
print("rmse:",rmse)
```

#### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=10,max_depth=5,random_state=42)

#fit the model
model.fit(X_train,y_train)

preds=model.predict(X_valid)

#score
mse= mean_squared_error(y_valid,preds)
r2=r2_score(y_valid,preds)
rmse= mean_squared_error(y_valid, preds)**0.5

print('mse ofRandomForestRegressor:',mse)
print("r2_score of RandomForestRegressor:",r2_score)
print("rmse of RandomForestRegressor:",rmse)

```

## HyperParameter Tuning 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Initialize the RandomForestRegressor
rf = RandomForestRegressor(random_state=42)

# Define the parameter grid
param_grid = {
    'n_estimators': [10],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, 
                                   n_iter=50, cv=5, verbose=2, n_jobs=-1, random_state=42)

# Fit the random search model
random_search.fit(X_train, y_train)

# View the best parameters
print(f"Best Parameters: {random_search.best_params_}")

# Predict on the validation set
best_model = random_search.best_estimator_
preds = best_model.predict(X_valid)

# Evaluate the model
rmse = np.sqrt(mean_squared_error(y_valid, preds))
r2 = r2_score(y_valid, preds)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R^2 Score: {r2}")
```

#### Usimg a simple imputer, creating a pipeline and hyperparameter tuning

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline



# Build a Pipeline for Data Preprocessing and Modeling

# Create a pipeline
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Handle missing values by filling with the mean
    ('scaler', StandardScaler()),  # Scale features to have zero mean and unit variance
    ('model', RandomForestRegressor(random_state=42))  # Random Forest model
])

#  Define the Parameter Grid

param_grid = {
    'model__n_estimators': [10],  # Increased number of trees for better accuracy
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4],
    'model__max_features': ['auto', 'sqrt', 'log2']
}

# Initialize RandomizedSearchCV

random_search = RandomizedSearchCV(estimator=pipeline, param_distributions=param_grid, 
                                   n_iter=50, cv=5, verbose=2, n_jobs=-1, random_state=42)

# Fit the Random Search Model

random_search.fit(X_train, y_train)

# View the Best Parameters

print(f"Best Parameters: {random_search.best_params_}")


#Predict on the Validation Set

best_model = random_search.best_estimator_
preds = best_model.predict(X_valid)

# Evaluate the Model

rmse = np.sqrt(mean_squared_error(y_valid, preds))
r2 = r2_score(y_valid, preds)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R^2 Score: {r2}")

```

## SVR

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
import numpy as np
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Scale features
    ('svr', SVR())  # Support Vector Regression model
])

# Step 4: Train the SVM model

pipeline.fit(X_train, y_train)

# Step 5: Predict on the Validation Set

preds = pipeline.predict(X_valid)

# Step 6: Evaluate the Model

rmse = np.sqrt(mean_squared_error(y_valid, preds))
r2 = r2_score(y_valid, preds)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R^2 Score: {r2}")
```

# Combining SVR and Random Forest Regressor using Averaging


```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)

# Train RandomForest and SVR models
rf_model = RandomForestRegressor(n_estimators=10, random_state=42)
svr_model = SVR(C=10, kernel='rbf', gamma='scale')

rf_model.fit(X_train, y_train)
svr_model.fit(X_train_scaled, y_train)

# Predict using both models
rf_preds = rf_model.predict(X_valid)
svr_preds = svr_model.predict(X_valid_scaled)

# Combine the predictions by averaging
final_preds = (rf_preds + svr_preds) / 2

# Evaluate the combined model
rmse = np.sqrt(mean_squared_error(y_valid, final_preds))
r2 = r2_score(y_valid, final_preds)

print(f"Combined Model RMSE: {rmse}")
print(f"Combined Model R^2 Score: {r2}")
```

# Catboost and RandomForest regressor using stacking 

```python
# Initialize the models
rf = RandomForestRegressor(random_state=42)
catboost = CatBoostRegressor(random_state=42, verbose=0)

# Train the models
rf.fit(X_train, y_train)
catboost.fit(X_train, y_train)

# Make predictions on the validation set
rf_valid_preds = rf.predict(X_valid)
catboost_valid_preds = catboost.predict(X_valid)

# Create a new dataset with the predictions as features
stacked_valid_features = np.column_stack((rf_valid_preds, catboost_valid_preds))

# Initialize and train the meta-model
meta_model = RidgeCV()
meta_model.fit(stacked_valid_features, y_valid)

# Evaluate the meta-model
stacked_valid_preds = meta_model.predict(stacked_valid_features)
stacked_rmse = np.sqrt(mean_squared_error(y_valid, stacked_valid_preds))
print(f"Stacked Model RMSE: {stacked_rmse}")
```

```python

from sklearn.model_selection import KFold# Initialize the RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
rf_param_grid = {
    'n_estimators': [100, 300, 500],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Perform RandomizedSearchCV
rf_random_search = RandomizedSearchCV(estimator=rf, param_distributions=rf_param_grid, 
                                      n_iter=30, cv=5, verbose=2, n_jobs=-1, random_state=42)
rf_random_search.fit(X_train, y_train)
best_rf = rf_random_search.best_estimator_

# Initialize the CatBoostRegressor
catboost = CatBoostRegressor(random_state=42, verbose=0)
catboost_param_grid = {
    'iterations': [200, 500, 700],
    'depth': [4, 6, 8, 10],
    'learning_rate': [0.01, 0.05, 0.1],
    'l2_leaf_reg': [1, 3, 5, 7]
}

# Perform RandomizedSearchCV
catboost_random_search = RandomizedSearchCV(estimator=catboost, param_distributions=catboost_param_grid, 
                                            n_iter=30, cv=5, verbose=2, n_jobs=-1, random_state=42)
catboost_random_search.fit(X_train, y_train)
best_catboost = catboost_random_search.best_estimator_

# Cross-Validation Predictions for Stacking
kf = KFold(n_splits=5, shuffle=True, random_state=42)

rf_oof_preds = np.zeros(len(X_train))
catboost_oof_preds = np.zeros(len(X_train))

for train_idx, valid_idx in kf.split(X_train):
    X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[valid_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[valid_idx]

    # Train models
    best_rf.fit(X_tr, y_tr)
    best_catboost.fit(X_tr, y_tr)

    # Predict out-of-fold
    rf_oof_preds[valid_idx] = best_rf.predict(X_val)
    catboost_oof_preds[valid_idx] = best_catboost.predict(X_val)

# Stacking with RidgeCV as Meta-Model
stacked_train_features = np.column_stack((rf_oof_preds, catboost_oof_preds))
meta_model = RidgeCV()
meta_model.fit(stacked_train_features, y_train)

# Validate the Stacking Model
rf_valid_preds = best_rf.predict(X_valid)
catboost_valid_preds = best_catboost.predict(X_valid)
stacked_valid_features = np.column_stack((rf_valid_preds, catboost_valid_preds))
stacked_valid_preds = meta_model.predict(stacked_valid_features)

# Evaluate RMSE
stacked_rmse = np.sqrt(mean_squared_error(y_valid, stacked_valid_preds))
print(f"Improved Stacked Model RMSE: {stacked_rmse}")
```
